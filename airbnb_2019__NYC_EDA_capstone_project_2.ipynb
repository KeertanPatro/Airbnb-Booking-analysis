## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Exploring zero price

zero_price=airbnb_df.loc[airbnb_df['price']==0]
zero_price.head()


````

The above data consists of all the listings which have zero price, there seems to be no reason why 0 is entered in the price column, so we replace it by mean of price corresponding to that neighbourhood_group

````python
# replacing 0 by mean corresponding to the neighbourhood group
airbnb_df['price']=airbnb_df.groupby('neighbourhood_group').apply(lambda x: x['price'].replace(0,x['price'].mean())).reset_index().sort_values(by='level_1').set_index('level_1')['price']
airbnb_df.describe()
````

````python
# Now let us look into Availibilty column
airbnb_df.loc[airbnb_df['availability_365']==0].head()
````

The above data shows the entries in which 'availibility_365' is zero, and from 'last_review' column we see that these listings are before 2019 so these listings might be out of buisness with airbnb or customers did not visit or did not review, and since we don't know the exact reason behind it we left it as it is.

````python
# Exploring Null values

# counting null values in each column
airbnb_df.isna().sum(axis=0)
````

We see that 'name', 'host_id', 'last_review', 'reviews_per_month' columns have null values, columns 'last_review', 'reviews_per_month' have large number of null values.

````python
# let us explore why these columns have large null values
airbnb_df.loc[airbnb_df['reviews_per_month'].isna()].head()
````

When we look at the data in which the columns 'reviews_per_month' has null values it is because number_of_reviews correponding to that is zero so we fill null values with zero

````python
# filling out null values in reviews_per_month with zero
airbnb_df['reviews_per_month'].fillna(0,inplace=True)
airbnb_df
````

#### Chart - 1

#### Chart - 2

Price distribution based on neighbourhood groups

````python
# Chart - 2 visualization code
# price distribution based on neighbourhood groups
manhatten=airbnb_df.loc[airbnb_df['neighbourhood_group']=='Manhattan']['price']
brooklyn=airbnb_df.loc[airbnb_df['neighbourhood_group']=='Brooklyn']['price']
bronx=airbnb_df.loc[airbnb_df['neighbourhood_group']=='Bronx']['price']
staten_island=airbnb_df.loc[airbnb_df['neighbourhood_group']=='Staten Island']['price']
queens=airbnb_df.loc[airbnb_df['neighbourhood_group']=='Queens']['price']

plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
(airbnb_df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)).plot.bar(color=['r','g','b','violet','y'])
plt.ylabel('Mean price',fontweight='bold')
plt.xticks(fontweight='bold')
plt.title('Mean price for each neighbourhood group',fontweight='bold')
plt.subplot(1,3,2)
plt.boxplot([manhatten,brooklyn,bronx,staten_island,queens],showfliers=False,labels=['Manhattan','Brooklyn','Bronx','Staten island','Queen'],
            patch_artist=True,boxprops={'linewidth':2,'color':'r','facecolor':'#e0e0e0'},
            whiskerprops = {'linewidth':2},capprops={'linewidth':2},
            medianprops={'linewidth':2,'color':'blue'}
           )
plt.xticks(fontweight='bold')
plt.ylabel('Price',fontweight='bold')
plt.title('Box plot for each neighbourhood group',fontweight='bold')
plt.subplot(1,3,3)
sns.violinplot(x='neighbourhood_group',y='price',data=airbnb_df.loc[airbnb_df['price']<400],showfliers=False)
plt.title('price distribution for each neighbourhood group',fontweight='bold')
plt.subplots_adjust(left=0.1,
                    bottom=0.2,
                    right=2,
                    top=0.9,
                    wspace=0.2,
                    hspace=0.3)
plt.show()
````

##### 1. Why did you pick the specific chart?

There are three plots in this chart, the first plot is mean price distribution for different neighbourhood groups, the second plot is box whisker plot showing 25%tile,50%tile(median),75%tile for different neighbourhood groups,the third plot is about distribution of price based on neighbourhood groups.

From the first plot we see that mean price of manhattan is highest compared to others, which suggests that manhattan is an expensive place, from box whisker plot we see that IQR for manhattan is high while for others it is less and from price distribution plot we see that there is a broad distribution of price range in manhattan while for others the distribution is not so broad this indicates that there are many listings with wide range of prices, hence there are lots of options and variety in choosing price range for the customers if they want a listing in manhattan.

#### Chart - 15 - Pair Plot 

````python
# Pair Plot visualization code
df=airbnb_df[['price','minimum_nights','number_of_reviews','availability_365']] # creating a data frame consisting of necessary numerical variables
df=df.loc[(df['price']<400) & (df['minimum_nights']<500)] # limiting value value of price and minimum nights to 75 %tile so that scale does not change because of outliers
sns.pairplot(df,diag_kind='kde')
plt.suptitle('Pair plots of various numerical variables',y=1.05,fontweight='bold')
plt.show()
````

##### 1. Why did you pick the specific chart?

We wanted  to see relation between various numerical variables and in general how the points are trending.For that we have used a pair plot.

##### 2. What is/are the insight(s) found from the chart?

As such it is difficult to find a general trend of points as the correlation is not strong enough so in general we see that the variables are not strongly correlated to each other. but some weak negative correlation can be found between price and number of reviews as we confirmed in correlation map.
